```python
import pandas as pd
from pymongo import MongoClient
import dash
from dash import dcc, html
import plotly.express as px
```

```python
url = 'https://raw.githubusercontent.com/nabilnehme/AAI_634O/0d8adc1d72abba6f06aad14ba6dd6f7c6e07f186/Week1/Sales.csv'

# Load the CSV from the raw URL into a DataFrame
df = pd.read_csv(url, encoding='ISO-8859-1')

# Create a new DataFrame with the selected columns
sales_data = df[['date', 'product_id', 'sales_amount', 'store_location']]

# Display the first few rows of the new DataFrame
sales_data.head()
```

```python
# Establish a connection to the MongoDB server
#client = MongoClient('mongodb://localhost:27017/')  # Connect to the MongoDB server (default localhost and port)
connection_string = "mongodb+srv://nabilnehme:nabilnehme@cluster0.51s8b.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

# Connect to the MongoDB Atlas cluster
client = MongoClient(connection_string)

# Create or access the 'sales_db' database
db = client['sales_db']

# Create or access the 'sales' collection
collection = db['sales']

# CLEAR COLLECTION BEFORE INSERTING
collection.delete_many({})  # Deletes all documents in the collection
print("Collection cleared successfully!")

# Convert the DataFrame to a list of dictionaries and insert into MongoDB
sales_data_dict = sales_data.to_dict(orient='records')
collection.insert_many(sales_data_dict)

print("Sales data inserted successfully into MongoDB!")
```

```python
# Retrieve and print the first 10 entries from the MongoDB collection
first_10_entries = collection.find().limit(10)

print("First 10 entries from MongoDB:")
for entry in first_10_entries:
    print(entry)
```

#Read: Query MongoDB to Extract Sales Data Based on Store Location, Date, or Product

Query by Store Location:

```python
# Query sales data by store location (e.g., 'Los Angeles')
sales_los_angeles = collection.find({'store_location': 'Los Angeles'})

print("Sales in Los Angeles:")
for entry in sales_los_angeles:
    print(entry)
```

Query by Date Range:

```python
# Query sales data by date range (e.g., sales after '2016-01-01')
sales_after_date = collection.find({'date': {'$gt': '2016-01-01'}})

print("Sales after 2016-01-01:")
for entry in sales_after_date:
    print(entry)
```

Query by Product ID:

```python
# Query sales data by product ID (e.g., 'FUR-BO-10001798')
sales_by_product = collection.find({'product_id': 'FUR-BO-10001798'})

print("Sales for product FUR-BO-10001798:")
for entry in sales_by_product:
    print(entry)
```

#Update: Modify Sales Amounts for Specific Records

```python
# Update sales amount for a specific product (e.g., increase sales for 'FUR-BO-10001798' by 10%)
collection.update_many(
    {'product_id': 'FUR-BO-10001798'},  # Filter by product_id
    {'$mul': {'sales_amount': 1.10}}     # Increase sales_amount by 10%
)

print("Sales amount for product FUR-BO-10001798 updated successfully!")
```

```python
# Query sales data by product ID (e.g., 'FUR-BO-10001798')
sales_by_product = collection.find({'product_id': 'FUR-BO-10001798'})

print("Sales for product FUR-BO-10001798:")
for entry in sales_by_product:
    print(entry)
```

#Delete: Remove Outdated Sales Records
Delete All Sales Records Before a Specific Date:

```python
# Delete all sales records before a certain date (e.g., before '2015-01-01')
collection.delete_many({'date': {'$lt': '2015-01-01'}})

print("Sales records before 2015-01-01 deleted successfully!")
```

#Visualize data from MongoDB using Dash

```python
# Query MongoDB to get total sales by store location
sales_by_location = collection.aggregate([
    {"$group": {"_id": "$store_location", "total_sales": {"$sum": "$sales_amount"}}},
    {"$sort": {"total_sales": -1}}  # Sort by total_sales in descending order
])

# Convert the results to a DataFrame for easier plotting
sales_data = pd.DataFrame(list(sales_by_location))

# Initialize the Dash app
app = dash.Dash(__name__)

# Create a bar chart using Plotly Express
fig = px.bar(sales_data, x='_id', y='total_sales', labels={'_id': 'Store Location', 'total_sales': 'Total Sales'},
             title='Total Sales by Store Location')

# Set up the layout of the app
app.layout = html.Div(children=[
    html.H1("Sales Data Visualization", style={'textAlign': 'center'}),
    dcc.Graph(id='sales-bar-chart', figure=fig)
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
